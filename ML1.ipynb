**Apple Quality**

## Источник дааных:
https://www.kaggle.com/datasets/tejpal123/apple-quality-analysis-dataset?resource=download

## Характер данных:
Реальные данные

## Описание задачи анализа: 
Этот набор данных содержит информацию о различных характеристиках фруктов, что позволяет получить представление об их свойствах. Набор данных включает такие сведения, как идентификатор фрукта, размер, вес, сладость, хрусткость, сочность, спелость, кислотность и качество.

## Детальное описание атрибутов

**A_id** - уникальный идентификатор яблока, служащий для однозначной идентификации каждого образца в датасете

**Size** - размер яблока в кубических сантиметрах, характеризующий физический объем фрукта

**Weight** - масса яблока в граммах, ключевой параметр для сортировки

**Sweetness** - уровень сладости яблока, отражающий содержание натуральных сахаров

**Crunchiness** - показатель хрусткости яблока, определяющий текстуру мякоти

**Juiciness** - степень сочности яблока, важная для оценки вкусовых качеств

**Ripeness** - уровень зрелости яблока ("ripe", "unripe", "overripe")

**Acidity** - показатель кислотности яблока, влияющий на баланс вкуса

**Quality** - итоговая оценка качества ("good" или "bad")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")

# 1. Загрузка данных
df = pd.read_csv('C:/Users/nbcom/Downloads/apple_store/apple_quality.csv')

# 2. Краткое описание датасета
print("Размерность датасета:", df.shape)
print("\nПервые 5 строк:")
display(df.head())

print("\nИнформация о датасете:")
df.info()

print("\nСтатистика по числовым атрибутам:")
display(df.describe())

print("\nКатегориальные атрибуты и примеры значений:")
for col in df.select_dtypes(include='object').columns:
    print(f"- {col}: {df[col].unique()[:5]} (всего уникальных: {df[col].nunique()})")

# 3. Предобработка и статистический анализ по каждому атрибуту

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"\nЧисловые колонки: {numeric_cols}")

for col in numeric_cols:
    print(f"\nАтрибут: {col}")
    mean_val = df[col].mean()
    std_val = df[col].std()
    missing = df[col].isnull().sum()
    print(f"Среднее значение: {mean_val:.3f}")
    print(f"Стандартное отклонение: {std_val:.3f}")
    print(f"Пропущенные значения: {missing}")
    
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f'Гистограмма {col}')
    plt.show()
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    print(f"Количество выбросов: {outliers.shape[0]}")

print("\nОбработка пропущенных значений:")
for col in numeric_cols:
    missing = df[col].isnull().sum()
    if missing > 0:
        median_val = df[col].median()
        df.loc[:, col] = df[col].fillna(median_val)
        print(f"Пропуски в {col} заполнены медианой: {median_val}")

# 4. Корреляционный анализ

if len(numeric_cols) > 0:
    print("\nКорреляционная матрица:")
    
    numeric_df = df[numeric_cols].copy()
    
    missing_after = numeric_df.isnull().sum().sum()
    print(f"Пропуски после обработки: {missing_after}")
    
    if missing_after == 0:
        corr_matrix = numeric_df.corr()
        display(corr_matrix)

        plt.figure(figsize=(12,10))
        sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, 
                   center=0, vmin=-1, vmax=1)
        plt.title("Тепловая карта корреляций между числовыми атрибутами")
        plt.tight_layout()
        plt.show()

        
        if len(numeric_cols) >= 2:
            selected_columns = numeric_cols[:min(5, len(numeric_cols))]
            print(f"\nСтроим парные графики для: {selected_columns}")
            
            
            if len(selected_columns) >= 2:
                sns.pairplot(numeric_df[selected_columns])
                plt.suptitle("Матрица графиков рассеяния выбранных признаков", y=1.02)
                plt.tight_layout()
                plt.show()
            else:
                print("Недостаточно числовых колонок для парного графика")
        else:
            print("Недостаточно числовых колонок для парного графика")
    else:
        print("Остались пропуски в данных. Невозможно построить корреляционную матрицу.")
else:
    print("В данных нет числовых колонок для корреляционного анализа")


# Дополнительная проверка качества данных
print("\n" + "="*50)
print("ИТОГОВАЯ ИНФОРМАЦИЯ О ДАННЫХ:")
print("="*50)
print(f"Общее количество строк: {df.shape[0]}")
print(f"Общее количество столбцов: {df.shape[1]}")
print(f"Числовые столбцы: {len(numeric_cols)}")
print(f"Общее количество пропусков после обработки: {df.isnull().sum().sum()}")
```
